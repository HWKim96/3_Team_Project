month train data 생성

```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('../pre_data/pre_train.csv', index_col=0)
train.head()
```

```python
# 타입 확인 -- 날짜가 object임
train.info()
```

```python
# date 타입 변경 이유는 key 값으로 묶어주기 위해서
train['date'] = pd.to_datetime(train['date'])
```

```python
month_train = train.groupby([pd.Grouper(key='date', freq='M'),'shop_id', 'item_id','date_block_num']).agg({"item_cnt_day":"sum"})
month_train.head()
```

```python
month_train = month_train.reset_index()
month_train.head()
```

```python
month_train = month_train.rename(columns={"item_cnt_day":"item_cnt_month"})
month_train = month_train.drop(columns="date")
month_train.head()
```

```python
month_train.head(10)
```

```python
month_train.to_csv("../group_data/month_train.csv")
```
