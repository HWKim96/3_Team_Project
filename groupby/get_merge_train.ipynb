```python
import pandas as pd
import numpy as np
```

```python
month_train = pd.read_csv("../group_data/month_train.csv", index_col=0)
pre_items = pd.read_csv("../pre_data/pre_items.csv", index_col=0)
pre_shops = pd.read_csv("../pre_data/pre_shops.csv", index_col=0)

pre_test = pd.read_csv("../pre_data/pre_test.csv", index_col=0)
```

```python
merge_train = pd.merge(pre_test, month_train, how='left', on=['shop_id', 'item_id', 'date_block_num'])
merge_train.head()
```

```python
merge_train['item_cnt_month'] = merge_train['item_cnt_month'].fillna(0).astype('int8')
merge_train.head()
```

```python
merge_train = pd.merge(merge_train, pre_items, how='left', on='item_id')
merge_train.head()
```

```python
merge_train = pd.merge(merge_train, pre_shops, how='left', on='shop_id')
merge_train.head()
```

```python
merge_train.info(show_counts=True)
```

```python
merge_train[merge_train['item_cnt_month'] < 0]
```

```python
# 극단값 제거
merge_train['item_cnt_month'] = merge_train['item_cnt_month'].clip(0, 20)
```

```python
merge_train.head()
```

```python
merge_train.info()
```

```python
def reduce_column_type(df):
    for col in df.columns:
        col_max = df[col].max()
        
        if col_max <= 127:
            df[col] = df[col].astype("int8")
        elif col_max <= 32767:
            df[col] = df[col].astype('int16')
        else:
            df[col] = df[col].astype('int32')
            
    return df
```

```python
merge_train.info()
```

```python
merge_train = reduce_column_type(merge_train)
merge_train.info(show_counts=True)
```

```python
merge_train.to_csv("../group_data/merge_train.csv")
```
